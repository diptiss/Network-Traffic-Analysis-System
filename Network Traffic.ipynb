```python
#Import pandas library for data loading and data analysis
import pandas as pd

#Import numpy library for numerical and mathematical operations
import numpy as np

#Import matplotlib library to create charts and visualizations
import matplotlib.pyplot as plt

#Import seaborn library for boxplot
import seaborn as sns

# Import scipy.stats library for hypothesis testing
import scipy.stats as stats
```

```python
#Load dataset from CSV file into DataFrame
df = pd.read_csv("ids.csv")

#Display last 50 rows of the dataset
print("Last 50 Records:")
df.tail(50)
```

```python
#Define required columns for analysis
columns = ['Destination Port', 'Flow Duration', 'Total Fwd Packets', 'Total Backward Packets', 
           'Total Length of Fwd Packets', 'Total Length of Bwd Packets', 'Fwd Packet Length Mean',
           'Bwd Packet Length Mean', 'Flow Bytes/s', 'Flow Packets/s', 'Packet Length Mean', 'Packet Length Std',
           'Average Packet Size', 'Active Mean', 'Idle Mean', 'Label']
#Create new dataframe with selected columns 
df_new = df[columns].copy()

#Displays new column in dataframe
print("Columns in new DataFrame:")
print(df_new.columns)

#Print toal number of columns in new DataFrame
print("Total number of columns:", df_new.shape[1])
```

```python
# Count frequency of each label category
label_counts = df_new['Label'].value_counts()

# Print label distribution
print("Label Distribution:")
print(label_counts)

# Print total number of records
print(f"\nTotal records: {len(df_new)}")

# Create bar chart
plt.figure(figsize=(7, 5))

# Plot bars with custom colors
bars = plt.bar(
    label_counts.index,
    label_counts.values,
    color=['steelblue', 'tomato'],
    edgecolor='black',
    width=0.5
)

# Display frequency values above each bar
for bar, val in zip(bars, label_counts.values):
    plt.text(
        bar.get_x() + bar.get_width()/2,
        bar.get_height() + 200,
        f'{val:,}',
        ha='center',
        va='bottom',
        fontsize=11,
        fontweight='bold'
    )

# Add chart title and axis labels
plt.title('Frequency of Each Label Category', fontsize=14)
plt.xlabel('Label', fontsize=12)
plt.ylabel('Frequency', fontsize=12) 

# Adjust x axis label size
plt.xticks(fontsize=11)

# Adjust layout for better spacing
plt.tight_layout()

# Display chart
plt.show()

# Calculate imbalance ratio
ratio = label_counts.max() / label_counts.min()

# Print dataset balance analysis
print(f"Balance Analysis")
print(f"Normal Traffic: {label_counts.get('Normal Traffic', 0):,}")
print(f"Attack        : {label_counts.get('Attack', 0):,}")
print(f"Imbalance ratio: {ratio:.2f}:1")

# Check whether dataset is balanced or imbalanced
if ratio > 1.5:
    print("\nConclusion: The dataset is imbalanced because Normal Traffic has more records than Attack.")
else:
    print("\nConclusion: The dataset is slightly imbalanced, but both classes are reasonably represented.")
```

```python
#Create figure for boxplot
plt.figure(figsize=(10, 9))

#Generate boxplot grouped by Label
sns.boxplot(
    x='Label',
    y='Fwd Packet Length Mean',
    data=df_new
)

#Apply logarithmic scale on y-axis
plt.yscale('log')

#Add chart title and labels
plt.title('Distribution of Fwd Packet Length Mean by Label', fontsize=14)
plt.xlabel('Traffic Category', fontsize=12)
plt.ylabel('Fwd Packet Length Mean', fontsize=12)

#Adjust layout
plt.tight_layout()

#Display plot
plt.show()
```

```python
# Define significance level
alpha = 0.05

# Get Flow Duration values for Normal Traffic
normal_flow = df_new[df_new['Label'] == 'Normal Traffic']['Flow Duration']

# Get Flow Duration values for Attack traffic
attack_flow = df_new[df_new['Label'] == 'Attack']['Flow Duration']

# Perform Independent Samples t-test
t_stat, p_value = stats.ttest_ind(
    normal_flow,
    attack_flow,
)

# Print hypotheses
print("Null Hypothesis (H₀):")
print("There is no significant difference in mean Flow Duration between")
print("Normal Traffic and Attack traffic.\n")

print("Alternative Hypothesis (H₁):")
print("There is a significant difference in mean Flow Duration between")
print("Normal Traffic and Attack traffic.\n")

# Print test result
print("T-test Result")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.6f}\n")

# Decision based on p-value
if p_value < alpha:
    print("Decision: Reject the Null Hypothesis (H₀)")
    print("Conclusion: Mean Flow Duration is significantly different")
    print("between Normal Traffic and Attack traffic.")
else:
    print("Decision: Fail to Reject the Null Hypothesis (H₀)")
    print("Conclusion: There is no significant difference in")
    print("mean Flow Duration between the two traffic categories.")
```
